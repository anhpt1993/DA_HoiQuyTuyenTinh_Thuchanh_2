# 0. Khai báo thư viện cần thiết

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.dummy import DummyRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import r2_score
```

# 1. Đọc dữ liệu từ file CSV

```python
df = pd.read_csv("https://s3.ap-southeast-1.amazonaws.com/learn.codegym.vn/media/public/2022/03/10171754/AmesHousing.csv", encoding = "ISO-8859-1")
df.head()
```

```python
df.info()
```

```python
df.describe()
```

# 2. Tiền xử lý dữ liệu

```python
# Do thuật toán hồi quy chỉ làm việc với dữ liệu dạng số nên ta chuyển các biến có kiểu Category thành biến "Dummy"
df1 = pd.get_dummies(df)
df1.head()
```

```python
df1.info()
```

```python
# Kiểm tra và liệt kê những cột bị thiếu dữ liệu
df1.columns[df1.isna().any()].tolist()

df1.isna().sum()
```

```python
# Thay thế giá trị khuyết thiếu bằng giá trị trung vị của cột tương ứng.
# Định nghĩa hàm impute_median() để thực hiện việc này
def impute_median(series):
    return series.fillna(series.median())

df1['Lot Frontage']= df1['Lot Frontage'].transform(impute_median)
df1['Mas Vnr Area']=df1['Mas Vnr Area'].transform(impute_median)
df1['BsmtFin SF 1']=df1['BsmtFin SF 1'].transform(impute_median)
df1['BsmtFin SF 2']=df1['BsmtFin SF 2'].transform(impute_median)
df1['Bsmt Unf SF']=df1['Bsmt Unf SF'].transform(impute_median)
df1['Total Bsmt SF']=df1['Total Bsmt SF'].transform(impute_median)
df1['Bsmt Full Bath']=df1['Bsmt Full Bath'].transform(impute_median)
df1['Bsmt Half Bath']=df1['Bsmt Half Bath'].transform(impute_median)
df1['Garage Cars']=df1['Garage Cars'].transform(impute_median)
df1['Garage Area']=df1['Garage Area'].transform(impute_median)
#Check remaining columns with NaN values
df1.columns[df1.isna().any()].tolist()
```

```python
# Loại bỏ cột "Garage Yr Blt"
df2 = df1.drop("Garage Yr Blt", axis = 1)
```

# 3. Xử lý dữ liệu

```python
# Đặt mảng y, chứa giá của các ngôi nhà tương ứng với cột "SalePrice"
y = df2.SalePrice.values

# Đặt mảng X chứa tất cả những cột còn lại
X = df2.drop("SalePrice", axis = 1).values
```

```python
# Kiểm tra kích thước X
X.shape
```

```python
# Kiểm tra kích thước y
y.shape
```

```python
# Dùng lệnh reshape để định dạng y thành vector cột
y = y.reshape(-1, 1)
y.shape
```

# 4. Huấn luyện mô hình

```python
# Chia bộ dữ liệu làm 2 phần: dữ liệu huấn luyện (train set) và dữ liệu kiểm tra (test set).
# Tham số test_size = 0.3 cho biết 30% dữ liệu được lấy để kiểm tra, 70% dữ liệu được dùng để huấn luyện mô hình

X_train, X_test,y_train, y_test= train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
# Huấn luyện mô hình
model_r = LinearRegression()
model_r.fit(X_train, y_train)
```

```python
# Kiểm tra model
scr = model_r.score(X_test, y_test)
scr
```

Nhận xét: Mô hình đạt dự báo chính xác ~ 90%

```python
# Sử dụng mô hình với bộ dữ liệu kiểm tra
y_pred = model_r.predict(X_test)
```

```python
# Đánh giá mô hình sử dụng r2-score, mean_squared_error & root_mean_squared_error
score = r2_score(y_test, y_pred)
print("r2_score is: ", score)
print("mean_squared_error is: ", mean_squared_error(y_test, y_pred))
print("root_mean_squared error is: ", np.sqrt(mean_squared_error(y_test, y_pred)))
```
